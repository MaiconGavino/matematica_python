#Modelo de Equação de Lotka-Volterra

##O Modelo de Equação de Lotka-Volterra, também conhecido por Modelo Presa-Predador é um par de equação diferenciais ordinárias, não lineares, de primeira ordem, que são utilizadas para descrever dinâmicas em sistemas bióligicos. 


O código abaixo define duas equações diferenciais ordinárias, a função definica como rk4 é definida por um método Runge-Kutta de quarta ordem a partir das condições informadas.


```
import numpy as np
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
```

```
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

def sis(t,x):
  vet = np.array([a*x[0]*(d-x[0])-b*x[0]*x[1]/(1+m*x[0]), epsilon*b*x[0]*x[1]/(1+m*x[0])-g*x[1] ])
  return vet

def rk4(t,x, h):
  k1 = h*sis(t,x)
  k2 = h*sis(t+h/2,x+k1/2)
  k3 = h*sis(t+h/2,x+k2/2)
  k4 = h*sis(t+h,x+k3)
  xn = x + k1/6.0 + k2/3.0 + k3/3.0 + k4/6.0
  return xn

epsilon = 0.90
for j in range(0, 10):
  h = 1.0
  t = [0.0]
  x = [np.array([1.2, 0.68])]
  n = int(1000/h)
  a = 0.5
  b = 2.5
  m = 7.0
  d = 1.1
  g = 0.26

  for i in range(0 , n):
    tn = t[0] + (i+j)*h
    xn = rk4(t[i], x[i], h)
    t.append(tn)
    x.append(xn)

  x = np.array(x)
  ax.plot(x[:,0], x[:,1], epsilon, zdir='z')
  ax.set_xlabel(r'$u$', fontsize = 18)
  ax.set_ylabel(r'$w$', fontsize = 18)
  ax.set_zlabel(r'$\varepsilon$', fontsize = 18)
  epsilon = (epsilon*100 + 1)/100

plt.show()
```
